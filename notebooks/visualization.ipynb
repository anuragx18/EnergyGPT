```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load datasets
df = pd.read_csv("feature_engineered_data.csv", parse_dates=["DateTime"])
df_daily = pd.read_csv("daily_energy_data.csv", parse_dates=["DateTime"])
df_hourly = pd.read_csv("hourly_energy_data.csv", parse_dates=["DateTime"])

print(df.head()) 
```

```python
plt.figure(figsize=(12, 5))
sns.lineplot(data=df_daily, x="DateTime", y="Total_Daily_Consumption")
plt.title("Daily Electricity Consumption Over Time")
plt.xlabel("Date")
plt.ylabel("Total Daily Consumption (kWh)")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(12, 5))
sns.lineplot(data=df_hourly, x="DateTime", y="Total_Hourly_Consumption")
plt.title("Hourly Electricity Consumption Over Time")
plt.xlabel("Date")
plt.ylabel("Total Hourly Consumption (kWh)")
plt.xticks(rotation=45)
plt.show()
```
